```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,"g+-")       #g =>green and + =>marker
```

```python
plt.plot(x,y,"rD") #r =>red and D=>Diamond
```

```python
plt.plot(x,y,color="red",marker="D")
```

```python
plt.plot(x,y,color="red",marker="D",linestyle='',markersize=20)
```

```python
plt.plot(x,y,color="red",alpha=0.9)   #alpha is ranging from 0 to 1 it indicates thickness of line
```
