```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_values=[1600,600,300,410,250]
exp_labels=["home rent","food","wifi","car",'other']
```

```python
plt.pie(exp_values,labels=exp_labels)
plt.axis
```

```python
plt.axis("equal")
plt.pie(exp_values,labels=exp_labels,radius=2,autopct="%0.1f%%",shadow=True,explode=[0,0.2,0,0,0],startangle=180)
plt.savefig("piechart.png",bbox_inches="tight",pad_inches=2,transparent=True)  #it will save piechart as image
```
