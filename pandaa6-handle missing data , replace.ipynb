```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("weather4.csv")
df
```

```python
new_df = df.replace([-800,-900,-780,-100,-200,-300],np.NaN)
new_df
```

```python
new_df = df.replace({
    'temprature' : [-800,-900,-780],
    'windspeed' : [-100,-200,-300],
    'event' : ['no event','0']
},np.NaN)
new_df
```

```python
new_df = df.replace({
    -800 : np.NaN,
    "no event" : "sunny"
})
new_df
```

```python
df = pd.read_csv("weather4.csv")
df
```

```python
new_df = df.replace('[A-Za-z]','',regex=True)
new_df
```

```python
df_1 = df.replace({
    'temprature' : '[A-Za-z]',
    'windspeed' : '[A-Za-z]'
},'',regex=True)
df_1
```
