```python
import pandas as pd
```

```python
df = pd.read_csv("weather.csv")
df
```

```python
df.pivot(index="day",columns="city")  #day as row and city as column
```

```python
#to get only temp values
df.pivot(index="day",columns="city",values="temprature") 
```

```python
df = pd.read_csv("weather5.csv")
df
```

```python
df.pivot_table(index="city",columns="day")
```

```python
df.pivot_table(index="city",columns="day",aggfunc="sum") #32+32=64,32+35=67
```

```python
df.pivot_table(index="city",columns="day",aggfunc="count")
```

```python
df.pivot_table(index="city",columns="day",aggfunc="mean")
```

```python
df.pivot_table(index="city",columns="day",margins=True)
```
