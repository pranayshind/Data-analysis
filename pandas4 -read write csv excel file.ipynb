```python
#supply dictionary for replace with na_values

'''

here temprature can be negative but windspeed could not be negative so
to replace -ve values of specific column like windspeed we can use dictionary

'''
df = pd.read_csv("weather2.csv", na_values={
    'temprature' : ["not avaliable", "n.a."],         #only convert not avaliable and n.a. values to Nan
    'windspeed' : ["not avaliable", "n.a.",-8,-7],
    'event': ["not avaliable", "n.a."]
})
df
```
