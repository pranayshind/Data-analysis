```python
#concating 2 dataframes horizontally and vertically  {axis=1 for vertical}
```

```python
import pandas as pd
```

```python
India_weather = pd.DataFrame({
    "city" : ["Mumbai","Delhi","Pune"],
    "temprature" : [34,36,32],
    "humidity" : [80,70,75]
})
India_weather
```

```python
US_weather = pd.DataFrame({
    "city" : ["LA","California","new york"],
    "temprature" : [22,24,20],
    "humidity" : [68,72,64]
})
US_weather
```

```python
df = pd.concat([India_weather,US_weather])
df
```

```python
df = pd.concat([India_weather,US_weather],ignore_index=True)
df
```

```python
df = pd.concat([India_weather,US_weather],keys=["India","US"])
df
```

```python
df.loc["India"]  #loc=>location
```

```python
df.loc["US"]
```

```python
temp = pd.DataFrame({
    "city" : ["Mumbai","Delhi","Pune"],
    "temprature" : [34,36,32],
})
temp
```

```python
humidity = pd.DataFrame({
    "city" : ["Mumbai","Delhi","Pune"],
    "humidity" : [80,70,75]
})
humidity
```

```python
df = pd.concat([temp,humidity])
df
```

```python
df = pd.concat([temp,humidity],axis=1)
df
```
